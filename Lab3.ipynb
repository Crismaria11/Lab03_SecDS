```python
from sklearn.preprocessing import StandardScaler
```

Recordar que si las columnas tienen dos valores es porque son categoricas 

```python
scaler = StandardScaler()
scalered = scaler.fit_transform(df[['new_kernel32', 'new_msvcrt', 'new_shell32', 'new_user32', 'new_ws232', 
                                    'new_ADVAPI32', 'new_WININET', 'new_ntdll']])
scalered

scaler2 = StandardScaler()
scalered2 = scaler2.fit_transform(df[['new_kernel32', 'new_msvcrt', 'new_shell32', 'new_user32', 'new_ws232', 
                                    'new_ADVAPI32', 'new_WININET', 'new_ntdll']])
scalered2
```

```python
from sklearn.cluster import KMeans
```

```python
Sum_of_squared_distances = []
K = range(1,16)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(scalered)
    Sum_of_squared_distances.append(km.inertia_)
```

```python
Sum_of_squared_distances
```

```python
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
wscc = []

for i in range(1, 16):
#     print(i)
    kmeans = KMeans(
                    init = "random",
                    n_clusters = i,
                    n_init = 1,
                    max_iter = 300,
                    random_state = 123
                    ).fit(scalered2)
    wscc.append(kmeans.inertia_)
wscc
```

```python
plt.plot(range(1,16), wscc, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_coefficients = []
for i in range(2, 16):
    kmeans = KMeans(init = "random",
                    n_clusters = i,
                    n_init = 1,
                    max_iter = 300,
                    random_state = 123
                    ).fit(scalered)
    score = silhouette_score(scalered, kmeans.labels_)
    silhouette_coefficients.append(score)
```

```python
plt.style.use("fivethirtyeight")
plt.plot(range(2, 16), silhouette_coefficients)
plt.xticks(range(2, 16))
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Coefficient")
plt.show()
```

```python
silhouette_coefficients = []
for i in range(2, 15):
    kmeans = KMeans(init = "random",
                    n_clusters = i,
                    n_init = 1,
                    max_iter = 300,
                    random_state = 123
                    ).fit(scalered2)
    score = silhouette_score(scalered2, kmeans.labels_)
    silhouette_coefficients.append(score)
```
